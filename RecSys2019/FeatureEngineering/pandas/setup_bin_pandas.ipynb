```python
%%time
# open file
# for rec in file
path = '/datasets/trivago/data/'
pd_df = pd.read_csv(f'{path}item_metadata.csv')
```

```python
%%time
pd_sr = pd_df['properties'].str.split('|', expand=True)
```

```python
%%time
new_sr = pd.Series()
for col in pd_sr.columns:
    new_sr = pd.concat([new_sr, pd_sr[col]], axis=0)
```

```python
%%time
cats = list(set(new_sr))
```

```python
%%time
cats.remove(None)
```

```python
len(cats)
```

```python
%%time
for col in cats:
    new_sr = pd_df['properties'].str.contains(col, regex=False)
    col_name = f'is_{col}'
    pd_df[col_name] = new_sr.astype('int')
```
